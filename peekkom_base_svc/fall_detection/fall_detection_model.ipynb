```python
import yaml

# 디렉토리 정보, 클래스 이름(names), 클래스 수(nc) 지정하기
data = { 'test': '/content/fall_detection/test',
        'train': '/content/fall_detection/train',
        'val': '/content/fall_detection/valid',
         'names' : ['-', 'fall', 'not_fallen'],
         'nc' : 3 }

# 데이터 경로와 클래스 정보를 저장하고 있는 딕셔너리 객체 data를 YOLOv8 학습에 필요한 새로운 이름으로 저장
with open('fall_Data.yaml', 'w') as f:
    yaml.dump(data, f)
```

```python
import pandas as pd
from ultralytics import YOLO
```

```python
model = YOLO('yolov8m.pt')
print(type(model.names), len(model.names))
print(model.names)
```

```python
model.train(data='fall_Data.yaml', epochs=50, patience=15, batch=16, imgsz=640, conf=0.3)
```

```python
print(type(model.names), len(model.names))

print(model.names)
```
